#### 2020&ensp;鼠年鼠战&ensp; Mickey Vs Jerry  


##### fast.ai深度学习框架在jetsonNano上快速搭建卷积神经网络，实现Mickey和Jerry的照片分类。

##### 从 fast.ai 读入一些相关的功能模块

```python
from fastai.vision import *
```

#### 1.准备数据（Data）

这里的任务是做分类。因此需要有标注的训练数据。

下面需要设置数据所在文件夹的位置，为 imgs 目录。

```python
path = Path('imgs')
```

下面，我们让 fast.ai 帮我们载入全部的数据。这时我们调用 ImageDataBunch 类的 from_folder 函数，结果存储到 data 中：


```python
# data = ImageDataBunch.from_folder(path, test='test', ds_tfms=get_transforms(), size=224)
data = ImageDataBunch.from_folder(path, test='test', size=224,bs=16)
```

这里，读入数据后对其进行处理和增强：

数据增强（augmentation），也就是对数据进行了翻转、拉伸、旋转，弄出了很多“新”训练数据。

这样做的目的，是因为数据越多，越不容易出现过拟合（over-fitting），也就是模型死记硬背，蒙混考试，却没有抓住真正的规律。

另外，把图片大小进行了统一，设置成了 224 x 224 ，这样做的原因，是我们需要使用迁移学习，要用到预训练模型。预训练模型是在这样大小的图片上面训练出来的，因此保持大小一致，训练效果更好。

接下来，检查一下机器学习的数据载入是否正常：（这里只显示出一部分，便于显示）

```python
data.show_batch(rows=3, figsize=(10,10))
```

##### 图片和标记都是正确的，再查看下类别数，训练集、验证集、测试集图片数量

```python
data.classes, data.c, len(data.train_ds), len(data.valid_ds),len(data.test_ds)
```

-------------------------------------

#### 2.创建模型结构（Architecture）开始训练

要让AI做图片分类，需要使用卷积神经网络（Convolutional Neural Network）来表征图片上的像素信息构成的特征。

用下面这一条语句，我们把“数据”、“模型结构”和“损失度量”三样信息，一起喂给机器。

```python
learn = cnn_learner(data, models.resnet18, metrics=accuracy)
```

数据就不说了，模型我们采用的是 resnet18 这样一个预训练模型作为基础架构。至于损失度量，我们用的是准确率（accuracy）。

这里，我们无需设置模型类别的个数，神经网络的层数、大小、激活函数等，

因为 fast.ai 自动根据你输入的上述“数据”、“模型结构”和“损失度量”信息，自动帮你把这些繁琐的操作搞定了。

下面，用fit指令来训练搭建好的卷积神经网络：

```python
learn.fit_one_cycle(1)
```

24秒钟之后，训练结束。

验证集准确率能达到90%左右。

注意，我们使用迁移学习“拿来”的这个 resnet18 模型当初做训练的时候，可从来没有见识过mickey和jerry。

神经网络学习了训练集200多张形态各异，包含各种背景噪声的图片，它就能 90% 准确分辨了。







-------------------------------------



#### 3.模型评估

用fastai解释工具进行训练效果损失度量（Loss Metrics）

下面创建一个模型解释工具，来衡量机器学习模型对某项任务的处理水平，这里解释工具用来评判模型的分类效果如何。

```python
interp = ClassificationInterpretation.from_learner(learn)
```

执行上面语句，创建解释工具。

接下来，通过解释器判断模型识别最差的9张图片

```python
interp.plot_top_losses(9, figsize=(10,10))
```

这个解释器可以让你明白，机器识别哪些图片信心不足。

接下来让解释器做个混淆矩阵出来：

```python
interp.plot_confusion_matrix()
```

横坐标是预测值，纵坐标是图片的真实标签，对角线的数字是正确推理的照片数量，其他区间是推理错误的图片数量。

##### 下面查看在训练集上的推理表现

图片上面的单词是真实标签，下面是模型的推理结果。

```python
learn.show_results(ds_type=DatasetType.Train,rows=4)
```

##### 查看在验证集上的推理表现

```python
learn.show_results(ds_type=DatasetType.Valid,rows=4)
```

我们刚才上面展示的，是验证集的结果。

这个验证集，机器在迭代模型参数的时候每一回都拿来尝试。

所以要评估模型的识别图片泛化能力，我们需要让模型看从来没有看过的测试集图片。

下面查看在测试集数据（data.test_ds）上的推理表现

```python
img = data.test_ds[9][0]
img
```

```python
learn.predict(img)
```

模型推理的输出，根据两个不同的分类，分别给出的倾向数值。数值越大，倾向程度越高。

如：tensor([0.3292, 0.6708]) 左侧一列，是jerry的确信度；右侧一列，是mickey的。

```python
data.classes
```

-------------------------------------

#### 4.微调（fine-tuning）

根据上面推理效果优化超参数继续训练模型

```python
learn.unfreeze()
learn.fit_one_cycle(1, slice(1e-5,3e-4))
```

补训练模型后，重复上面第3步的评估步骤。

以上就是一个卷积神经网络的标准训练流程。
